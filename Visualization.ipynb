## T_c vs Rho_c

```python
l1 = mr.MesaLogDir('1Mass/LOGS/')
l2 = mr.MesaLogDir('2Mass/LOGS/')

Prof_nums1=l1.profile_numbers
Prof_nums2=l2.profile_numbers
```

```python
Central_Ts1=np.zeros_like(Prof_nums1)
Central_Rhos1=np.zeros_like(Prof_nums1)

Central_Ts2=np.zeros_like(Prof_nums2)
Central_Rhos2=np.zeros_like(Prof_nums2)

for i,Prof in enumerate(Prof_nums1):
    n = l1.profile_data(profile_number=Prof)
    Central_Ts1[i]=n.logT[-1]
    Central_Rhos1[i]=n.logRho[-1]
    
for i,Prof in enumerate(Prof_nums2):
    n = l2.profile_data(profile_number=Prof)
    Central_Ts2[i]=n.logT[-1]
    Central_Rhos2[i]=n.logRho[-1]
```

```python
plt.plot(Central_Ts1,Central_Rhos1)
plt.plot(Central_Ts2,Central_Rhos2)
```
